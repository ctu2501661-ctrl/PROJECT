```
# LOADING THE DATA SET
import pandas as pd
df = pd.read_csv("primevideo_dataset (1).csv")
print(df)
```

```
import matplotlib.pyplot as plt

#1 MOVIES VS TV DURATION
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```
import matplotlib.pyplot as plt

#2. Content Type Share (Pie Chart)
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Content Type Share')
plt.ylabel('')
plt.show()
```

```
import matplotlib.pyplot as plt

#3 3. Release Year Distribution
df['release_year'].plot(kind='hist', bins=10)
plt.title('Release Year Distribution')
plt.xlabel('Year')
plt.show()
```

```
import matplotlib.pyplot as plt

#4. Scatter Plot

#What it shows: Relationship between release year and dataset entries
plt.scatter(df['release_year'], df.index)
plt.title('Scatter Plot of Release Year vs Index')
plt.xlabel('Release Year')
plt.ylabel('Index')
plt.show()
```

```
#5. Line Graph

#What it shows: Trend of content released over the years
df.groupby('release_year').size().plot(kind='line')
plt.title('Line Graph of Content Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Shows')
plt.show()
```

```
#6. Horizontal Bar Chart

#What it shows: Country-wise distribution of content
df['country'].value_counts().plot(kind='barh')
plt.title('Horizontal Bar Chart of Content by Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```
#7. Pie Chart

#What it shows: Percentage share of movies and TV shows
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Content Type')
plt.ylabel('')
plt.show()
```

```
#8 Histogram

#What it shows: Distribution of content across release years

df['release_year'].plot(kind='hist', bins=10)
plt.title('Histogram of Release Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```
#9. Release Year Trend
df.groupby('release_year').size().plot(title='Release Trend')
```

```
#10.Genre Distribution Pie Chart
df['listed_in'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
```

```
#11 . Box Plot of Movie Duration

# Filter for movies
movies = df[df['type'] == 'Movie'].copy()

# Extract numerical duration and convert to int for movies
movies['duration_min'] = movies['duration'].str.extract('(\d+)').astype(int)

# Create the box plot
movies.boxplot(column='duration_min')
```

```
#12 TOP ACTORS
df['cast'].value_counts().head(5).plot(kind='bar', title='Top Actors')
```

```
#13 TOP DIRECTORS
df['director'].value_counts().head(5).plot(kind='bar', title='Top Directors')
```

```
#14Content Added Over Time
df['date_added'] = pd.to_datetime(df['date_added'])
df['date_added'].dt.year.value_counts().sort_index().plot(title='Content Added Per Year')
```

```
#15Duration Analysis (Movies Only)
movies = df[df['type'] == 'Movie']
movies['duration_min'] = movies['duration'].str.replace(' min','').astype(int)
movies['duration_min'].plot(kind='hist', bins=10, title='Movie Duration')
```
